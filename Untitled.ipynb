## Deleting features with low correlation

```python
def drop_columns(df, correlation_threshold):
    target_param = 'Median_House_Value'
    correlations = df.corr()[target_param]
    
    columns_to_drop = []
    for column, correlation in correlations.items():
        if abs(correlation) < correlation_threshold and column != target_param:
            columns_to_drop.append(column)

    df_temp = df.copy()
    df_temp = df_temp.drop(columns=columns_to_drop)
    return df_temp
```

```python
target_param = 'Median_House_Value'
df.shape
```

```python
predictor_columns = df.columns[~df.columns.isin(['Median_House_Value'])]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error, r2_score
```

## Value of alpha is not too decisive in case of r^2 and mse

```python
def perform_ridge_regression(predictor_columns, df, target_column, test_size=0.1):
    
    results_list = []
    X = df[predictor_columns]
    y = df[target_column]

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
    alpha=.01

    for i in range(100):
        model = Ridge(alpha=alpha)
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
    
        results = {
            'Alpha': alpha,
            'MSE': mse,
            'R^2': r2,
            'Coefficients': model.coef_
        }
    
        results_list.append(results)
            
        alpha+=0.01
    

    return results_list
```

```python
# bedzie mozna w petli posprawdzac wartosci mse dla roznych: alpha, testsize, i roznych wartosci korelacji
```

```python
results = perform_ridge_regression(predictor_columns, df, target_param)
```

```python
alphas = [result['Alpha'] for result in results]
r2_values = [result['R^2'] for result in results]
mse_values = [result['MSE'] for result in results]
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.plot(alphas, mse_values, marker='o', linestyle='-', color='b')
plt.title('WartoÅ›ci mse w zaleÅ¼noÅ›ci od Alpha')
plt.xlabel('Alpha')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```

## Now let's test how dependant results are on feature amount

### To check it we remove features from dataset based on their correlation

```python
def perform_ridge_regression_changeFeatures(df, target_column, test_size=0.1):
    results_list = []
    alpha=.1
    model = Ridge(alpha=alpha)
    correlation_threshold = 0.01

    for i in range(50):
        df_temp = drop_columns(df, correlation_threshold)
        predictor_columns = df_temp.columns[~df_temp.columns.isin(['Median_House_Value'])]
        X = df_temp[predictor_columns]
        y = df_temp[target_column]
    
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size)
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
    
        mse = mean_squared_error(y_test, y_pred)
        r2 = r2_score(y_test, y_pred)
    
        results = {
            'Correlation': correlation_threshold,
            'MSE': mse,
            'R^2': r2,
            'Coefficients': model.coef_
        }
        correlation_threshold += 0.01
    
        results_list.append(results)
    
    return results_list 
```

```python
results_list = perform_ridge_regression_changeFeatures(df, target_param)
r2 = [result['R^2'] for result in results_list]
correlation = [result['Correlation'] for result in results_list]
mse = [result['MSE'] for result in results_list]
```

## We can see that in this case not removing any data results in a better model

```python
plt.figure(figsize=(10, 6))
plt.plot(correlation, r2, marker='o', linestyle='-', color='b')
plt.title('Value of r^2 with respect to correlation')
plt.xlabel('Correlation')
plt.ylabel('R^2')
plt.grid(True)
plt.show()
```

## On the other hand we see that removing one feature might result in lower MSE

```python
plt.figure(figsize=(10, 6))
plt.plot(correlation, mse, marker='o', linestyle='-', color='b')
plt.title('Value of mse with respect to correlation')
plt.xlabel('Correlation')
plt.ylabel('MSE')
plt.grid(True)
plt.show()
```
