### Now we will drop columns connected with case count because that's what we are trying to predict, we also delete INCOMPLETE

```python
columns_to_drop = ['BX_CASE_COUNT', 'BX_PROBABLE_CASE_COUNT', 'BK_CASE_COUNT', 'BK_PROBABLE_CASE_COUNT', 'MN_CASE_COUNT', 'MN_PROBABLE_CASE_COUNT', 'QN_CASE_COUNT', 'QN_PROBABLE_CASE_COUNT', 'SI_CASE_COUNT', 'SI_PROBABLE_CASE_COUNT', 'INCOMPLETE']
df = df.drop(columns_to_drop, axis=1)
```

```python
df = df.set_index('date_of_interest')
```

```python
df.corr()
```

### We will delete PROBABLE_CASE_COUNT because there is no information about this column in the documentation

```python
df = df.drop('PROBABLE_CASE_COUNT', axis=1)
```

```python
df['DEATH_COUNT'].max()
```

```python
sns.lineplot(data=df, x='date_of_interest', y='CASE_COUNT')
sns.lineplot(data=df, x='date_of_interest', y='DEATH_COUNT')
```

```python
from sklearn.linear_model import Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = df.drop(columns=['CASE_COUNT'])
y = df['CASE_COUNT']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = Ridge(alpha=.1)
```

```python
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

r2 = r2_score(y_pred, y_test)
r2
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train.copy())
X_test_scaled = scaler.transform(X_test.copy())

model.fit(X_train_scaled, y_train)
y_pred = model.predict(X_test_scaled)

r2 = r2_score(y_pred, y_test)
r2
```

```python
selector = RFE(estimator=model, n_features_to_select=30)
```

```python
selector.fit(X_train.copy(), y_train.copy())

selected_features_train = selector.transform(X_train.copy())
selected_features_test = selector.transform(X_test.copy())

model.fit(selected_features_train, y_train)
y_pred = model.predict(selected_features_test)
r2 = r2_score(y_pred, y_test)
r2
```

```python

selector.fit(X_train_scaled.copy(), y_train.copy())

selected_features_train = selector.transform(X_train_scaled.copy())
selected_features_test = selector.transform(X_test_scaled.copy())

model.fit(selected_features_train, y_train)
y_pred = model.predict(selected_features_test)
r2 = r2_score(y_pred, y_test)
r2
```
