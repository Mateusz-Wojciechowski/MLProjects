```python
import pandas as pd
import numpy as np
import matplotlib as ma
```

```python
df = pd.read_csv('Salary_dataset.csv')
```

```python
df
```

```python
df.shape
df.index = df["Unnamed: 0"]
df = df.drop("Unnamed: 0", axis=1)
df
```

```python
X = df[['YearsExperience']]
Y = df['Salary']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.4)
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
import matplotlib.pyplot as plt
```

```python
mae = mean_absolute_error(y_test, y_pred)
mae
```

```python
plt.scatter(X_test, y_test, label='Rzeczywiste dane testowe')
plt.plot(X_test, y_pred, color='red', label='Przewidywane wartości')
plt.legend()
```

```python
differences = y_test - y_pred
differences
```

```python
x_indices = np.arange(len(y_test))

# Tworzenie wykresu słupkowego
plt.bar(x_indices, differences)

# Dodawanie tytułu i etykiet osi
plt.title('Różnice między przewidywanymi a rzeczywistymi wartościami')
plt.xlabel('Numer punktu')
plt.ylabel('Różnica')
```
